```python
import matplotlib.pyplot as plt
import networkx as nx
```

## Testing the networkx package

```python
G = nx.Graph()
# adding a node
G.add_node(1)
#adding a list of nodes with attributes
G.add_nodes_from([((2,2),{"color":"green"})
                 ,(3,{"color":"yellow"})
                 ,(2,{})])
#adding a list of edges with weight
G.add_weighted_edges_from([(1,(2,2),2)])

#drawing the graph
nx.draw(G, with_labels=True,font_weight='bold')
plt.show()
```

## Running Dikjstra using Networkx

```python
G = nx.karate_club_graph()

pos = nx.kamada_kawai_layout(G)

path = nx.shortest_path(G,source=14,target=16,weight=1.)

path_edges = []
for i in range(len(path)-1):
    edge = (path[i],path[i+1])
    path_edges.append(edge)

print(path_edges)

print(path)
nx.draw(G,pos,node_color="black",with_labels=True)
nx.draw_networkx_nodes(G,pos,nodelist=path,node_color='r')
nx.draw_networkx_edges(G,pos,edgelist=path_edges,edge_color='r',width=3)
plt.show()
```
