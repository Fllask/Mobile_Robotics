```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display

from Utilities import Utilities
ut = Utilities()
from Global import Global
import matplotlib.pyplot as plt

import numpy as np
```

```python
def displayGlobal(sx=80.,sy=20.,fx=20.,fy=70.):
    start = (sx,sy)
    finish = (fx,fy)

    test = Global(ut.TestMap(),start,finish)

    test.plotPath()
    test.netPlot(test.navGraph)
```

```python
w = interactive(displayGlobal,sx=(0.,100.),sy=(0.,100.),fx=(0.,100.),fy=(0.,100.))
display(w)
```

```python
def displayNetwork(sx=80.,sy=20.,fx=20.,fy=70.):
    start = (sx,sy)
    finish = (fx,fy)

    test = Global(ut.TestMap(),start,finish)

    test.netPlot(test.computeGraph())
```

```python
w = interactive(displayNetwork,sx=(0.,100.),sy=(0.,100.),fx=(0.,100.),fy=(0.,100.))
display(w)
```
